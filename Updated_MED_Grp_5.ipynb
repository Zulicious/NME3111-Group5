```python
import pandas as pd
import matplotlib.pyplot as plt

#Load the CSV file
data = pd.read_csv('VesselActivity.csv')

#Convert 'Date' and 'Time' columns to a single datetime column
data['Datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d/%m/%Y %I:%M %p')

#Extract the date from the datetime column
data['Date'] = data['Datetime'].dt.date

#Filter out rows where the engine is OFF (assuming speed is 0 when OFF)
data = data[data['E'] == 'ON']

#Calculate the average speed per day
daily_speed = data.groupby('Date')['Speed'].mean().reset_index()

#Plotting the daily average speed
plt.figure(figsize=(12, 6))
plt.plot(daily_speed['Date'], daily_speed['Speed'], marker='o')
plt.title('Daily Average Vessel Speed')
plt.xlabel('Date')
plt.ylabel('Speed (knots)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Print the daily average speed data
print(daily_speed)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("VesselTripHistory.csv")

# Try multiple datetime formats to correctly parse 'Start Time' and 'Stop Time'
date_formats = ["%Y-%m-%d %H:%M:%S", "%d/%m/%Y %H:%M", "%m/%d/%Y %I:%M %p"]
for fmt in date_formats:
    try:
        df["Start Time"] = pd.to_datetime(df["Start Time"], format=fmt, errors='coerce')
        df["Stop Time"] = pd.to_datetime(df["Stop Time"], format=fmt, errors='coerce')
        if df["Start Time"].notna().all() and df["Stop Time"].notna().all():
            break  # Exit loop if successful
    except:
        continue

# Filter rows where vessel is running
df_run = df[df["Run/Idle"].str.lower() == "run"].copy()

# Drop rows with missing 'Stop Time' or 'Start Time'
df_run = df_run.dropna(subset=["Start Time", "Stop Time"])

# Ensure Stop Time is always after Start Time
df_run = df_run[df_run["Stop Time"] > df_run["Start Time"]]

# Calculate total distance covered
total_distance = df_run["Distance (nm)"].sum()

# Calculate daily average speed
daily_avg_speed = df_run.groupby(df_run["Start Time"].dt.date)["Avg Speed"].mean()

# Find max speed recorded
max_speed = df_run["Max Speed"].max()

# Calculate total operating hours
df_run["Operating Hours"] = (df_run["Stop Time"] - df_run["Start Time"]).dt.total_seconds() / 3600
total_operating_hours = df_run["Operating Hours"].sum()
total_hours = int(total_operating_hours // 1)
total_minutes = int((total_operating_hours % 1) * 60)

# Load the CSV file
trip_df = pd.read_csv("VesselTripHistory.csv")

# Convert the "Duration" column (formatted as hh:mm:ss) into timedelta format
trip_df["Duration_td"] = pd.to_timedelta(trip_df["Duration"], errors='coerce')

# Convert durations to minutes
trip_df["Duration_min"] = trip_df["Duration_td"].dt.total_seconds() / 60

# Filter only the trips where the vessel was running (not idling)
run_trips = trip_df[trip_df["Run/Idle"].str.lower() == "run"]

# Sum all run durations and convert to hours
total_run_minutes = run_trips["Duration_min"].sum()
total_run_hours = total_run_minutes / 60

# Plot total operational hours
plt.figure(figsize=(4, 5))
plt.bar(["Total Operational Hours for a Month"], [total_run_hours], color='purple', alpha=0.7)
plt.ylabel("Hours")
plt.title("Total Operational Hours for Run Trips")
plt.ylim(0, total_run_hours + 5)
plt.text(-0.52, total_run_hours, f'{total_run_hours:.2f} h', ha='center', va='bottom', fontsize=7, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



# Print results

print(f"Total Operational Hours (run trips only): {total_run_hours:.2f} hours")
```

```python
import pandas as pd
import numpy as np

# Load CSV
trip_df = pd.read_csv('VesselTripHistory.csv')

# Convert 'Duration' to timedelta and calculate duration in minutes
trip_df['Duration_td'] = pd.to_timedelta(trip_df['Duration'], errors='coerce')
trip_df['Duration_min'] = trip_df['Duration_td'].dt.total_seconds() / 60

# STEP 3: Filter only 'run' trips with valid duration, distance, and speed
trip_df['Distance (nm)'] = pd.to_numeric(trip_df['Distance (nm)'], errors='coerce')
trip_df['Avg Speed'] = pd.to_numeric(trip_df['Avg Speed'], errors='coerce')

run_trips = trip_df[
    (trip_df['Run/Idle'] == 'run') &
    (trip_df['Duration_min'].notna()) &
    (trip_df['Distance (nm)'].notna()) &
    (trip_df['Avg Speed'].notna())
].copy()

# Calculate total operational time
total_run_hours = run_trips['Duration_min'].sum() / 60
print(f"🕒 Total Operational Time (run trips): {total_run_hours:.1f} hours")

# STEP 4: Create Distance Range Table
distance_bins = [0, 10, 20, 30, 40, 50]
distance_labels = ['0–10 nm', '10–20 nm', '20–30 nm', '30–40 nm', '40–50 nm']
run_trips['Distance Range'] = pd.cut(run_trips['Distance (nm)'], bins=distance_bins, labels=distance_labels, right=False)

distance_distribution = run_trips['Distance Range'].value_counts().sort_index()
distance_table = pd.DataFrame({
    'Distance Range': distance_distribution.index,
    'Number of Trips': distance_distribution.values
})
print("\n📏 Distance Range Distribution (nautical miles):")
display(distance_table)

# STEP 5: Create Speed Range Table
speed_bins = [0, 5, 10, 15, 20, 25, 30]
speed_labels = ['0–5 kn', '5–10 kn', '10–15 kn', '15–20 kn', '20–25 kn', '25–30 kn']
run_trips['Speed Range'] = pd.cut(run_trips['Avg Speed'], bins=speed_bins, labels=speed_labels, right=False)

speed_distribution = run_trips['Speed Range'].value_counts().sort_index()
speed_table = pd.DataFrame({
    'Speed Range': speed_distribution.index,
    'Number of Trips': speed_distribution.values
})
print("🚤 Speed Range Distribution (knots):")
display(speed_table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # Import the numpy library

activity_df = pd.read_csv('VesselActivity.csv')
activity_df = activity_df[activity_df['E'] == 'ON'].copy()

# STEP 3: Define vessel speed-power curves and auxiliary loads
vessel_designs = {
    'Design 1: Composite (High Speed)': {0: 0,8.0: 23, 8.8: 31, 12.5: 85, 15.0: 120, 18.5: 170},
    'Design 2: Composite (Normal Speed)': {0: 0, 8.0: 22, 8.8: 30, 12.5: 80, 15.0: 110},
    'Design 3: Aluminium (Large)': {0: 0, 10: 120, 11: 176, 12: 245, 13: 300},
    'Design 4: Aluminium (Small)': {0: 0,3: 1.6, 5: 5, 8: 30.6, 10: 73.6, 12: 138.8, 14: 204.4},
    'Design 5: Hydrofoil': {0: 0, 2: 0, 4: 2, 6: 10, 8: 23, 10: 41, 12: 65, 14: 89,
                            16: 108, 18: 114, 20: 120, 22: 133, 24: 151}
}

aux_loads = {
    'Design 1: Composite (High Speed)': 10,
    'Design 2: Composite (Normal Speed)': 10,
    'Design 3: Aluminium (Large)': 18,
    'Design 4: Aluminium (Small)': 10,
    'Design 5: Hydrofoil': 50
}

# STEP 4: Calculate energy and cost
electricity_cost_per_kWh = 0.55
total_minutes = len(activity_df)

results = []

for design_name, sp_curve in vessel_designs.items():
    speeds = np.array(list(sp_curve.keys()))
    powers = np.array(list(sp_curve.values()))

    # Vectorized power interpolation
    interpolated_power = np.interp(activity_df['Speed'], speeds, powers)
    energy_kWh = interpolated_power / 60  # kWh per minute

    total_prop_kWh = energy_kWh.sum()
    aux_kWh = aux_loads[design_name] * total_minutes / 60
    total_kWh = total_prop_kWh + aux_kWh
    total_cost = total_kWh * electricity_cost_per_kWh

    results.append({
        'Design': design_name,
        'Propulsion Energy (kWh)': round(total_prop_kWh, 2),
        'Auxiliary Energy (kWh)': round(aux_kWh, 2),
        'Total Energy (kWh)': round(total_kWh, 2),
        'Monthly Energy Cost (SGD)': round(total_cost, 2)
    })

# STEP 5: Display results
results_df = pd.DataFrame(results)
print("\n🔋 Monthly Energy and Cost Summary from Vessel Activity:")
display(results_df)
```

```python
import pandas as pd
import numpy as np

# Load CSV
trip_df = pd.read_csv('VesselTripHistory.csv')
trip_df['Duration_td'] = pd.to_timedelta(trip_df['Duration'], errors='coerce')
trip_df['Duration_hr'] = trip_df['Duration_td'].dt.total_seconds() / 3600

# Filter valid 'run' trips
trip_df_clean = trip_df[
    (trip_df['Run/Idle'] == 'run') &
    (trip_df['Avg Speed'].notna()) &
    (trip_df['Duration_hr'].notna())
].copy()

# Speed-power curves and auxiliary loads
vessel_designs = {
    'Design 1: Composite (High Speed)': {8.0: 23, 8.8: 31, 12.5: 85, 15.0: 120, 18.5: 170},
    'Design 2: Composite (Normal Speed)': {0.0: 0, 8.0: 22, 8.8: 30, 12.5: 80, 15.0: 110},
    'Design 3: Aluminium (Large)': {0: 0, 10: 120, 11: 176, 12: 245, 13: 300},
    'Design 4: Aluminium (Small)': {3: 1.6, 5: 5, 8: 30.6, 10: 73.6, 12: 138.8, 14: 204.4},
    'Design 5: Hydrofoil': {
        0: 0, 2: 0, 4: 2, 6: 10, 8: 23, 10: 41, 12: 65,
        14: 89, 16: 108, 18: 114, 20: 120, 22: 133, 24: 151
    }
}

aux_loads = {
    'Design 1: Composite (High Speed)': 10,
    'Design 2: Composite (Normal Speed)': 10,
    'Design 3: Aluminium (Large)': 18,
    'Design 4: Aluminium (Small)': 10,
    'Design 5: Hydrofoil': 50
}

# STEP 3: Compute max energy per design
results = []

for design, sp_curve in vessel_designs.items():
    speeds = np.array(list(sp_curve.keys()))
    powers = np.array(list(sp_curve.values()))
    aux = aux_loads[design]

    df = trip_df_clean.copy()
    df['Propulsion Power (kW)'] = np.interp(df['Avg Speed'], speeds, powers)
    df['Trip Energy (kWh)'] = (df['Propulsion Power (kW)'] + aux) * df['Duration_hr']

    max_trip = df.loc[df['Trip Energy (kWh)'].idxmax()]

    results.append({
        'Design': design,
        'Max Energy (kWh)': round(max_trip['Trip Energy (kWh)'], 2),
        'Max Trip Duration (hr)': round(max_trip['Duration_hr'], 2),
        'Max Trip Speed (kn)': round(max_trip['Avg Speed'], 2)
    })

# STEP 4: Show and optionally save
energy_summary = pd.DataFrame(results)
print("🔋 Max Energy Demand Summary (Per Design):")
display(energy_summary)
```
